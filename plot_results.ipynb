```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_results(file, name):
    df = pd.read_csv(file)
    runs = df['top1_val'].groupby(df.index // 10).apply(list) # a list of lists where each list is a run
    average_accuracy = df.groupby('iteration')['top1_val'].mean() # a list where each element is the average accuracy per iteration

    iterations = range(10)
    # plt.figure(figsize=(10, 6))

    # Plot each run
    for run in runs:
        plt.plot(iterations, run, linestyle='-')

    # Plot trendline
    coefficients = np.polyfit(iterations, average_accuracy, 2)
    trend_line = np.poly1d(coefficients)
    plt.plot(iterations, trend_line(iterations), linestyle='--', label=f'Trend Line (slope: {coefficients[0]:.4f})')

    plt.xticks(range(10))
    plt.xlabel('Iteration')
    plt.ylabel('Accuracy')
    plt.title(f'{name}: Accuracies and Trend Line over Iterations')
    # plt.legend()
    plt.show()
    return trend_line(iterations)
```

Least confidence sampling

```python
trend_least = plot_results('Results/fashiontop40_results_least.csv', 'Least Confidence')
```

Margin sampling

```python
trend_margin = plot_results('Results/fashiontop40_results_margin.csv', 'Margin')
```

Entropy sampling

```python
trend_entropy = plot_results('Results/fashiontop40_results_entropy.csv', 'Entropy')
```

Random sampling

```python
trend_random = plot_results('Results/fashiontop40_results_random.csv', 'Random')
```

Traditional ML

```python
trend_traditional= plot_results('Results/fashiontop40_results_noAL.csv', 'Traditional ML')
```

All five trendlines in one plot


```python
iterations = range(10)

trend_lines = {"Least": trend_least, "Margin": trend_margin, "Entropy": trend_entropy, "Random": trend_random, "Traditional": trend_traditional}

for name, trend_line in trend_lines.items():
    plt.plot(iterations, trend_line, linestyle='--', label=f'{name}')
plt.xticks(range(10))
plt.yticks()
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Trend Lines over Iterations')
plt.legend()
plt.show()
```

Increase of accuracy from iteration 0

```python
import matplotlib.pyplot as plt

iterations = range(10)
trend_lines = {"Least": trend_least, "Margin": trend_margin, "Entropy": trend_entropy, "Random": trend_random, "Traditional": trend_traditional}

# Normalizing trend lines
normalized_trend_lines = {}
for name, trend_line in trend_lines.items():
    normalized_trend_line = [y - trend_line[0] for y in trend_line]
    normalized_trend_lines[name] = normalized_trend_line

# Plotting trend lines
for name, trend_line in normalized_trend_lines.items():
    plt.plot(iterations, trend_line, linestyle='--', label=f'{name}')
plt.xticks(range(10))
plt.xlabel('Iteration')
plt.ylabel('Change in Accuracy')
plt.title('Increase of Accuracy over Iterations')
plt.legend()
plt.show()
```
